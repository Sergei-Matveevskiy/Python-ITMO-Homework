**Работа с API**

```python
import requests
import pandas as pd
```

```python
# Получение книг
response_books = requests.get("https://www.anapioficeandfire.com/api/books")
books = response_books.json()
books_df = pd.DataFrame(books)
books_df.head()
```

```python
# Получение домов
response_houses = requests.get("https://www.anapioficeandfire.com/api/houses")
houses = response_houses.json()
houses_df = pd.DataFrame(houses)
houses_df.head()
```

```python
# Получение домов с девизом
houses_with_words = [house for house in houses if 'words' in house and house['words']]
houses_with_words_df = pd.DataFrame(houses_with_words)
houses_with_words_df.head()
```
